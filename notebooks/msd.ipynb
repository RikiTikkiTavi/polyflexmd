```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def msd_adj(t, a, R, tau_R):
    return a * R * (1 + np.exp(-(t-100)**2 / (2*40**2))/100 - np.exp(-t / tau_R))

def msd_adj_bonded(t, a, R, tau_R, tau_B):
    return a * R * (1 - (np.exp(-t / tau_R ) + np.exp(- 3**2 * t / tau_R ) + np.exp(- 5**2 * t / tau_R ))/3)


ax: plt.Axes
fig, ax = plt.subplots(figsize=(10,10))

R = 0.7
a = 1
tau_R = 20_000
ts = np.linspace(0, 10*tau_R, 10*tau_R)

ax.plot(ts, msd_adj(ts, a, R, tau_R), label="adjusted")
ax.plot(ts, msd_adj_bonded(ts, a, R*1.3, tau_R, tau_R), label="adjusted bonded")

ax.legend()
ax.set(xscale="log", yscale="log")
```
