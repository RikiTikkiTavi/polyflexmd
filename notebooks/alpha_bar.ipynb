```python
import matplotlib.pyplot as plt
import pandas as pd

import seaborn as sns
import polyflexmd.data_analysis.plot.utils as plot_utils

%load_ext autoreload
%autoreload 2

sns.set_style("darkgrid")

sns.set_style("darkgrid")

params = {
    'axes.labelsize': 10,
    'figure.dpi': 300,
    "figure.figsize": plot_utils.get_figure_size(width=1, n_rows=1)
}

plt.rcParams.update(params)
```

```python

alpha_min = pd.DataFrame([
    ("a", 0.73, 0.02),
    ("b", 0.67, 0.02),
    ("c", 0.652, 0.002),
    ("d", 0.640, 0.002),
], columns=["case", "alpha", "delta alpha"]).set_index("case")

alpha_min
```

```python
ax: plt.Axes

fig, ax = plt.subplots()

ax.bar(alpha_min.index, alpha_min["alpha"], label="$\\alpha_{min}$")

ax.set(ylabel="$\\alpha_{min}$", ylim=(0.5, 0.775))
ax.errorbar(x=alpha_min.index, y=alpha_min["alpha"], yerr=alpha_min["delta alpha"], linestyle="", color=sns.color_palette()[1], label="$\\alpha_{min}$ CI")

ax.set_yticks([0.50, 0.60, 2/3, 0.7, 3/4])
ax.set_yticklabels([0.50, 0.60, "2/3", 0.7, "3/4"])

ax.axhline(y=3/4, color="black", linestyle="--", alpha=0.7)
ax.axhline(y=2/3, color="black", linestyle="--", alpha=0.7)

ax.legend()
```

```python
delta_alpha = pd.DataFrame([
    ("$a \\rightarrow b$", 0.06, 0.03),
    ("$c \\rightarrow d$", 0.012, 0.003),
    ("$a \\rightarrow c$", 0.08, 0.02),
    ("$a \\rightarrow d$", 0.09, 0.02),
], columns=["case", "d_alpha", "delta d_alpha"]).set_index("case")

delta_alpha
```

```python
ax: plt.Axes

fig, ax = plt.subplots()

ax.bar(delta_alpha.index, delta_alpha["d_alpha"], label="$\\Delta \\alpha_{min}$")

ax.set(ylabel="$\\Delta \\alpha_{min}$", )
ax.errorbar(
    x=delta_alpha.index, y=delta_alpha["d_alpha"], yerr=delta_alpha["delta d_alpha"], linestyle="", color=sns.color_palette()[1],
    label="$\\Delta \\alpha_{min}$ CI"
)

ax.axhline(y=0.083, linestyle="--", color="black", alpha=0.7, label="$\\Delta \\alpha_{min}$ by Singh et al.")

ax.legend()
```
