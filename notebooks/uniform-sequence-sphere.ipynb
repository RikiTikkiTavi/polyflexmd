```python
import numpy as np
```

```python
N_bonds = 127
N_walks = 10000
bond_length=.97
```

```python
def get_walk(N: int, bond_length: float):
    phis = np.random.uniform(low=0, high=2 * np.pi, size=N)
    thetas = np.random.uniform(low=0, high=np.pi, size=N)
    rs = bond_length * np.vstack([
        np.sin(thetas) * np.cos(phis),
        np.sin(thetas) * np.sin(phis),
        np.cos(thetas)
    ]).T
    return rs

walks = [get_walk(N=N_bonds, bond_length=bond_length) for _ in range(N_walks)]
walks[0]
```

```python
def calculate_ete_vec(walk: np.ndarray) -> np.ndarray:
    return np.sum(walk, axis=0)

ete_vecs = [calculate_ete_vec(w) for w in walks]
ete_vecs
```

```python
np.sum(ete_vecs, axis=0) / len(ete_vecs)
```
